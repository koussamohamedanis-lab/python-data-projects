```python
import numpy as np

from sklearn.linear_model import LinearRegression

 

# Exemple de données d'entraînement

irradiance = np.array([1000, 900, 800, 1100, 1050]).reshape(-1, 1)

temperature = np.array([25, 23, 22, 27, 26]).reshape(-1, 1)

production = np.array([300, 270, 240, 330, 315])

 

# Combiner les données

X = np.hstack((irradiance, temperature))

 

# Modèle de régression linéaire

model = LinearRegression()

model.fit(X, production)

 

# Prédiction

new_data = np.array([[1000, 24]])

predicted_production = model.predict(new_data)

print(f"Production prédite: {predicted_production[0]:.2f} kWh")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
 
x=np.arange(1,30,0.1)
y=np.sin(x)
y2=np.cos(x)
plt.plot(x,y,color='red', label='sin')
plt.plot(x,y2,color='g', label='cos')
plt.xlabel("x-axs")
plt.ylabel("y-axs")
plt.title("sin")
plt.legend()
plt.grid(True)
plt.show()
```
