```python
import numpy as np

# Define the coefficients of the polynomial
coeff = [2, -3, -12, 5]

# Creating the polynomial
poly = np.poly1d(coeff)

# Find the derivative of the polynomial
deriv = np.polyder(poly)
print(deriv)

# Find the roots of the derivative to find the critical points
crit_pts = np.roots(deriv)
print(crit_pts)

# Filter out any complex roots (if any)
real_roots = crit_pts[np.isreal(crit_pts)].real

# Evaluate the original polynomial at the real roots to get the extrema
extrema_x = real_roots
extrema_y = poly(extrema_x)

# Print the extrema coordinates
print("Extrema X-coordinates:", extrema_x)
print("Extrema Y-coordinates:", extrema_y)

# Plot the polynomial to check the results
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 100)
plt.plot(x, poly(x))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y=2x^3 -3x^2 - 12x + 5')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data points
x = np.array([1, 2, 3, 4, 5])
w = np.array([2, 3, 5, 7, 11])

# Fit a quadratic polynomial to the data
coeffs = np.polyfit(x, w, 4)

# Create a polynomial function using the coefficients
poly_function = np.poly1d(coeffs)
print (poly_function)
# Generate values for x to plot the polynomial curve
x_values = np.linspace(0, 6, 100)
y_values = poly_function(x_values)

# Plot the original data points
plt.scatter(x, w, color='blue', label='Data Points')

# Plot the fitted polynomial curve
plt.plot(x_values, y_values, color='red', label='Quadratic Fit')

plt.xlabel('x')
plt.ylabel('w')
plt.title('Quadratic Polynomial Fit')
plt.legend()
plt.grid(True)
plt.show()
```
