```python
import tweepy
import pandas as pd

# Set up your Twitter API credentials
consumer_key = 'your_consumer_key'
consumer_secret = 'your_consumer_secret'
access_token = 'your_access_token'
access_token_secret = 'your_access_token_secret'

# Authenticate with the Twitter API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)

# Search for tweets containing the keyword 'Python'
query = 'Python'
try:
    # Use the updated method for searching tweets
    tweets = api.search_tweets(q=query, lang='en', count=10)
    
    # Extract tweet data
    tweet_data = [{'text': tweet.text, 'created_at': tweet.created_at} for tweet in tweets]
    df = pd.DataFrame(tweet_data)
    print(df.head())

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import nltk
nltk.download('movie_reviews')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk import PorterStemmer
from nltk.corpus import movie_reviews
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag, ne_chunk


documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]


porter = PorterStemmer()


print(documents[0])
print(documents[1])

text = ' '.join(documents[1][0])


print(sent_tokenize(text))


textnig = word_tokenize(text)


for token in textnig[10:]:
    print(porter.stem(token))


tagged = pos_tag(textnig)


entities = ne_chunk(tagged)

for entity in entities[10:]:
    print(entity)
```

```python
import spacy


nlp = spacy.load('en_core_web_sm')


text = "The quick brown fox jumps over the lazy dog."


doc = nlp(text)

tokens = [token.text for token in doc]
print("Tokens:", tokens)


lemmas = [token.lemma_ for token in doc]
print("Lemmas:", lemmas)

pos_tags = [(token.text, token.pos_) for token in doc]
print("POS Tags:", pos_tags)


entities = [(ent.text, ent.label_) for ent in doc.ents]
print("Named Entities:", entities)
for token in doc:
    print (token, token.idx)
```
