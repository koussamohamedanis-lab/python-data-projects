```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(9, 9))

# Plot the data on the subplots
axs[0, 0].plot(x, y1,label='sin(x)',color='green')
axs[0, 0].set_title('Sine Plot')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].legend()

axs[0, 1].plot(x, y2,label='cos(x)',color='cyan')
axs[0, 1].set_title('Cosine plot')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].legend()

axs[1, 0].plot(x, y3,label='tan(x)',color='red')
axs[1, 0].set_title('Tangent Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].legend()

axs[1, 1].plot(x, y4,label='sin(x)',color='yellow')
axs[1, 1].set_title('Exponential plot')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].legend()

# Adjust layout to prevent overlap of labels
plt.tight_layout()

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 20)
y = 2 * x**2 - 3 * x + 1
z = np.random.randint(0, 10, size=(20))
w = z + y
coeffs = np.polyfit(x, w, 2)
p_fit = np.poly1d(coeffs)
print(p_fit)
plt.scatter(x, w, label='data')
xfit = np.linspace(-5, 5, 100) 
yfit = np.polyval(coeffs, xfit)
plt.plot(xfit, yfit, label='fit')
plt.legend()
plt.show()
```
